## AVERAGES

```python
Ageavg = df['Age'].mean(skipna = True)
Expavg = df['Years of experience'].mean(skipna = True)
Salavg = df['Yearly Salary'].mean(skipna = True)
print(Ageavg,Expavg,Salavg)
```

```python
df['Age'] = df['Age'].fillna('Not mentioned')
```

```python
import numpy as np
df["Number of vacation days"].replace({"Not mentioned": np.nan}, inplace=True)
```
